# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### Object oriented api method

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.9,0.9]) # left, bottom, width and height of axis 
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes=plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')


plt.tight_layout()
```

### Figure Size and DPI

```python
fig=plt.figure(figsize=(3,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(figsize=(8,2))
# ax=fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))
# ax=fig.add_axes([0,0,1,1])
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))
# ax=fig.add_axes([0,0,1,1])
axes[0].plot(x,y)
axes[1].plot(y,x)
# plt.tight_layout()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='square')
ax.plot(x,x**3,label='cube')

# ax.legend(loc=0)
ax.legend(loc=(0.1,0.1))
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue')
ax.plot(y,x,color='red')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue', linewidth=2,alpha=0.5)
ax.plot(y,x,color='red',linewidth=2)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue', linewidth=2,alpha=0.5)
ax.plot(y,x,color='red',lw=2,ls=':')
```
