```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator='std')
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator='mean')
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips) #quartiles
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker') #quartiles
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex') #quartiles
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
sns.violinplot(x='day',y='total_bill',data=tips,color='black')
```
